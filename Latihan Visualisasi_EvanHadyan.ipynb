```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data = pd.read_csv('covid_19_indonesia_time_series_all.csv')
data.head()
```

```python
data.info()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.info()
```

```python
data['Month-Year'] = data['Date'].dt.to_period('M')
data.head()
```

```python
data['Month-Year'].nunique()
```

```python
data['Month-Year'].value_counts()
```

```python
grouped = data.groupby('Month-Year')[['Total Deaths', 'Total Recovered', 'Total Active Cases']].median().reset_index()
grouped #reset_index agar month-year berubah jadi fitur
```

```python
grouped['Month-Year'] = grouped['Month-Year'].astype('str')
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.title('Deaths, Recovered, and Active Cases of Covid 2019')
plt.xlabel('Year-Month')
plt.ylabel('Victims')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'])
plt.plot(grouped['Month-Year'], grouped['Total Recovered'])
plt.plot(grouped['Month-Year'], grouped['Total Active Cases'])
plt.title('Deaths, Recovered, and Active Cases of Covid 2019')
plt.xlabel('Year-Month')
plt.ylabel('Victims')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(grouped['Month-Year'], grouped['Total Deaths'],
         label = 'Total Deaths')
plt.plot(grouped['Month-Year'], grouped['Total Recovered'],
         label = 'Total Recovered')
plt.plot(grouped['Month-Year'], grouped['Total Active Cases'],
         label = 'Total Active Cases')
plt.title('Deaths, Recovered, and Active Cases of Covid 2019')
plt.xlabel('Year-Month')
plt.ylabel('Victims')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
#membuat 2 fig & 2 ax
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(grouped['Month-Year'], grouped['Total Recovered'],
         label = 'Total Recovered', color='#B9F3FC', linestyle='--')
ax1.set_title('Recovered Cases of Covid 2019')
ax1.set_xlabel('Year-Month')
ax1.set_ylabel('Number Of Recovered')
ax1.tick_params(axis = 'x', rotation = 45)
ax1.legend()


ax2.plot(grouped['Month-Year'], grouped['Total Deaths'],
         label = 'Total Deaths', color='#301E67', linestyle='dotted')
ax2.plot(grouped['Month-Year'], grouped['Total Active Cases'],
         label = 'Total Active Cases', color='#FF8B13', linestyle='dashdot')
ax2.set_title('Active Covid and death')
ax2.set_xlabel('Year-Month')
ax2.set_ylabel('Death and Active Cases')
ax2.tick_params(axis = 'x', rotation = 45)
ax2.legend()


plt.tight_layout()

plt.show()

fig1.savefig('fig1.png', bbox_inches='tight') #bbox agar gambar tidak terpotong
fig2.savefig('fig2.png', bbox_inches='tight')
```
